```python
x = list(gap_data.keys())
y = list(gap_data.values())

plt.title('Gap Data')
plt.plot(x, y, 'bo-')
plt.xlabel('Number of measurements between peaks')
plt.ylabel(r'$\Delta$ Speed')

plt.axhline(y=340, color='k', linestyle='-', lw=1)
plt.axvline(x=6.69, color='k', linestyle='-', lw=1)

plt.axhline(y=350, color='m', linestyle='-', lw=.5)
plt.axvline(x=6.5, color='m', linestyle='-', lw=.5)

plt.axhline(y=330, color='m', linestyle='-', lw=.5)
plt.axvline(x=6.89, color='m', linestyle='-', lw=.5)

plt.show()

#estas son las velocidades obtenidas para en función del tamaño del gap (número de mediciones del arduino entre picos).
#Nótese que a medida que aumenta el tamaño del gap, las diferencias entre las velocidades es cada vez menor, por lo que el valor de la velocidad calculada tiende a converger
#dado un número fijo de pulsos emitidos en cierta frecuencia (en este caso, 90 y 90).
```
